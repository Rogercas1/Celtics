```python
url = "https://www.basketball-reference.com/players/t/tatumja01/gamelog/2022"
```

```python
import requests
data = requests.get(url)
```

```python
with open("Jayson.html", "w+") as f:
        f.write(data.text)
```

```python
from bs4 import BeautifulSoup
```

```python
with open("Jayson.html") as f:
    page = f.read()
```

```python
soup = BeautifulSoup(page, "html.parser")
```

```python
Tatum = soup.find(id="pgl_basic")
```

```python
import pandas as pd
```

```python
df = pd.read_html(str(Tatum))[0]
```

```python
df
```

```python
df.drop('Rk', axis=1, inplace=True)
```

```python
df.drop('Age', axis = 1, inplace=True)
```

```python
df
```

```python
df.drop(df.index[(df["G"] == "G")],axis=0,inplace=True)
```

```python
df.drop(df.index[(df["GS"] == "Inactive")],axis=0,inplace=True)
```

```python
df
```

```python
df.rename(columns = {'Unnamed: 5':'H or A'}, inplace = True)
```

```python
df
```

```python
df.rename(columns = {'Unnamed: 7':'Result'}, inplace = True)
```

```python
df
```

```python
df['H or A'] = df['H or A'].replace(['@'],'A')
```

```python
import numpy as np
```

```python
df['H or A'] = df['H or A'].replace([np.nan],'H')
```
